```python
RGB_IMG = cv2.imread('/content/color_template.png')
RGB_IMG = cv2.cvtColor(RGB_IMG, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(RGB_IMG)
height, width = RGB_IMG.shape[:2]
print(width,height)
```

```python
YIQ_IMG = RGB_IMG.copy()
RGBTOYIQ = [[0.299, 0.587, 0.114], [0.5959, -0.2746, -0.3213], [0.2115, -0.5227, 0.3112]]

#RGB to YIQ Dönüşümü
for x in range(419):
  for y in range(559):
    RGB = RGB_IMG[x,y,:]
    YIQ_IMG[x,y,:] = np.matmul(RGBTOYIQ,RGB)
    
plt.imshow(YIQ_IMG)
```

```python
#YIQ TO RGB Dönüşümü

YIQTORGB = [[1, 0.956, 0.619], [1, -0.272, -0.647], [1, -1.106, 1.703]]

for x in range(419):
  for y in range(559):
    YIQ = YIQ_IMG[x,y,:]
    RGB_IMG[x,y,:] = np.matmul(YIQTORGB,YIQ)

plt.imshow(RGB_IMG)
    
```

```python
#resize 50 percent
print('Original Dimensions : ',cat1.shape)

scale_percent = 50  
width = int(cat1.shape[1] * scale_percent / 100)
height = int(cat1.shape[0] * scale_percent / 100)
dim = (width, height)

resized = cv2.resize(cat1, dim, interpolation = cv2.INTER_AREA)
 
print('Resized Dimensions : ',resized.shape)
plt.imshow(resized)
```
